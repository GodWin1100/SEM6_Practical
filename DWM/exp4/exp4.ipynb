```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('iris.csv')
```

```python
print('Data\n',df.head())
print('\nshape :',df.shape)
```

```python
lr=LinearRegression()
x,y=df[['sepalwidth']],df['sepallength']
lr.fit(x,y)
print(lr.intercept_)
print(lr.coef_)
y_pred=lr.coef_+lr.intercept_*x
print(lr.score(x,y))
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,s=10)
plt.xlabel('Sepal Wdith')
plt.ylabel('Sepal Length')
plt.plot(x, y_pred, color = "g")
plt.show()
```
